# 4. Fill in all the missing values present in all the columns in the dataset

Age,Cabin and Embarked has null values

## Age

```python
print('\nNumber of null values in column Age ', df['Age'].isnull().sum())
```

Trying to fill the missing values with median 

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

## Embarked

The column 'Embarked'  has only 2 null values, which can be filled with mode.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```
