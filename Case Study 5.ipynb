## Total sales generated by all the zones for each month

```python
df['TotalSales'] = df.sum(axis=1)
max_value = df['TotalSales']. max()
min_value=df['TotalSales'].min()
max_index=df['TotalSales'].idxmax()
min_index=df['TotalSales'].idxmin()
```

## Check whether all the zones generate the same amount of sales.

```python
import scipy.stats as stats

stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

Since pvalue is smaller than alpha value, we reject the null hypothesis

That means amount of sales generated by each zone are different
